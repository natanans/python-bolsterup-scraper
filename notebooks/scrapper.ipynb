```python
import sqlite3
import json

def query_products():
    db_name = 'products.db'  # Name of the database
    conn = sqlite3.connect('../products.db')
    cursor = conn.cursor()

    # Example query: Fetch all records from the `products` table
    cursor.execute('SELECT * FROM products')

    # Fetch all results
    results = cursor.fetchall()

    # Print column headers
    headers = [description[0] for description in cursor.description]
    print(headers)

    # Print each row
    for row in results:
        # Convert JSON fields back to Python objects for readability
        product = {
            'id': row[0],
            'url': row[1],
            'detailed_description': row[2],
            'consumption_information': row[3],
            'image_url': row[4],
            'documents': json.loads(row[5]),  # Convert JSON string to Python dict
            'categories': json.loads(row[6])  # Convert JSON string to Python dict
        }
        print(product)

    # Close the connection
    conn.close()

query_products()
```
